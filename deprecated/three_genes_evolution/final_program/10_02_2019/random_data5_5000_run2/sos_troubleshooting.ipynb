```python
import pandas as pd
```

```python
def rearrange_file(file):
    """
    Rearranges filesfile[protein_species] so that the target input file can be compared to the simulated output file so that sum of squares can be easily calculated
    """
    protein_species = ['proteinX', 'proteinY', 'proteinZ']
    file = file[file['species'].isin(protein_species)]
    file = file[['time', 'species', 'transcript']]
    file['time'] = file['time'].round().astype(int)
    file = file.pivot(index='time', columns='species', values='transcript')
    file = file.fillna(0.0)
    for protein in protein_species:
        if protein not in file.columns:
            file[protein] = 0.0
    file = file[protein_species]
    return file
```

```python
df = pd.read_csv("three_genes_replicated_180.tsv", header=0, sep="\t")
df = rearrange_file(df)
df1 = pd.read_csv("../../random_data5.tsv", header=0, sep='\t')
df1 = rearrange_file(df1)
```
