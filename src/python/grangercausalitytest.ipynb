```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd

import sys

sys.path.insert(1, './lib/')

import file_setup
import yaml

from scipy import stats
```

```python
with open('gene_best.yml', 'r') as gene_parameters:
        genome_tracker_new = yaml.safe_load(gene_parameters)
data1 = file_setup.rearrange_file(pd.read_csv("positive_control2.tsv", header=0, sep='\t'), genome_tracker_new)
data2 = file_setup.rearrange_file(pd.read_csv("expression_pattern_best.tsv", header=0, sep='\t'), genome_tracker_new)

#Use for normal
# data = pd.concat([data1['protein1'], data2['protein1']], axis=1)
# data.columns = ['protein1_target','protein1_sim']

# # data1['protein1'].plot()
# # data2['protein1'].plot()

# data1['protein2'].plot()
# data2['protein2'].plot()
# print(data1.head())
# print(data2.head())

# #Use for differencing
data1_diff_1 = []
data2_diff_1 = []

data1_diff_2 = []
data2_diff_2 = []

data1_diff_3 = []
data2_diff_3 = []

for i in range(1, len(data1)):
    data1_diff_1.append((data1.iloc[i]['protein1'] - data1.iloc[i-1]['protein1']))
    data2_diff_1.append((data2.iloc[i]['protein1'] - data2.iloc[i-1]['protein1']))
    
    data1_diff_2.append((data1.iloc[i]['protein2'] - data1.iloc[i-1]['protein2']))
    data2_diff_2.append((data2.iloc[i]['protein2'] - data2.iloc[i-1]['protein2']))
    
    data1_diff_3.append((data1.iloc[i]['protein3'] - data1.iloc[i-1]['protein3']))
    data2_diff_3.append((data2.iloc[i]['protein3'] - data2.iloc[i-1]['protein3']))
    
data1_diff = pd.DataFrame(data=zip(data1_diff_1, data1_diff_2, data1_diff_3), columns=["protein1_diff", "protein2_diff", "protein3_diff"])
data2_diff = pd.DataFrame(data=zip(data2_diff_1, data2_diff_2, data2_diff_3), columns=["protein1_diff", "protein2_diff", "protein3_diff"])
```
